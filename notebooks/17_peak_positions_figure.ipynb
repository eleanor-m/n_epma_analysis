Summarise results for buddingtonite

```python
with open('../data/interim/peak_positions/buddingtonite_StA/buddingtonite.pickle', 'rb') as f:
    data = pickle.load(f)

peak_pos = []
kraw_pcnt = []
kraw_stdev_pcnt = []
raw_cps = []
raw_cps_stdev = []
comment = []
date = []

for d in data:
    peak_pos.append(d.corrected.loc[d.corrected['element'] == 'N', 'pos'].values[0])
    kraw_pcnt.append(d.montecarlo.loc[d.montecarlo['element'] == 'N', 'kraw_pcnt'].values[0])
    kraw_stdev_pcnt.append(d.montecarlo.loc[d.montecarlo['element'] == 'N', 'kraw_stdev_pcnt'].values[0])
    raw_cps.append(d.peak.loc[2, 'raw_cps'])
    raw_cps_stdev.append(d.peak.loc[2, 'stdev_raw_cps'])
    comment.append(d.info.comment)
    date.append(d.info.date)
    
df_budd = pd.DataFrame({'peak_pos': peak_pos, 
                'kraw_pcnt': kraw_pcnt, 'kraw_stdev_pcnt': kraw_stdev_pcnt, 
                'raw_cps': raw_cps, 'raw_cps_stdev': raw_cps_stdev,
                'comment': comment, 'date': date})
df_budd.sort_values('peak_pos')
```

```python
with open('../data/interim/peak_positions/D2872_StA/D2872.pickle', 'rb') as f:
    data = pickle.load(f)

peak_pos = []
kraw_pcnt = []
kraw_stdev_pcnt = []
raw_cps = []
raw_cps_stdev = []
comment = []
date = []

for d in data:
    peak_pos.append(d.corrected.loc[d.corrected['element'] == 'N', 'pos'].values[0])
    kraw_pcnt.append(d.montecarlo.loc[d.montecarlo['element'] == 'N', 'kraw_pcnt'].values[0])
    kraw_stdev_pcnt.append(d.montecarlo.loc[d.montecarlo['element'] == 'N', 'kraw_stdev_pcnt'].values[0])
    raw_cps.append(d.peak.loc[0, 'raw_cps'])
    raw_cps_stdev.append(d.peak.loc[0, 'stdev_raw_cps'])
    comment.append(d.info.comment)
    date.append(d.info.date)
    
df_glass = pd.DataFrame({'peak_pos': peak_pos, 
                'kraw_pcnt': kraw_pcnt, 'kraw_stdev_pcnt': kraw_stdev_pcnt, 
                'raw_cps': raw_cps, 'raw_cps_stdev': raw_cps_stdev,
                'comment': comment, 'date': date})
df_glass.sort_values('peak_pos')
```

```python
def calc_means(df):
    df_means = df.groupby('peak_pos').mean()[['kraw_pcnt', 'raw_cps']]
    df_means = df_means.rename(columns={nm: nm + '_mean' for nm in df_means.columns})
    df_stdev = df.groupby('peak_pos').std()[['kraw_pcnt', 'raw_cps']]
    df_stdev = df_stdev.rename(columns={nm: nm + '_stdev' for nm in df_stdev.columns})
    df_means = pd.concat([df_means, df_stdev], axis=1)
    return df_means
```

```python
df_budd_means = calc_means(df_budd)
df_glass_means = calc_means(df_glass)
```

## UP TO HERE : RECREATING THE FIGURE!

```python
fig, axs = plt.subplots(ncols=1, nrows=3, figsize = (8*0.39, 18*0.39), sharex=True)
titlefontsize = 10
# Panel A: raw cps for glass
ax = axs[0]
ax.set_ylabel("raw cps")
ax.set_title("Glass D2872", fontsize=titlefontsize, loc="left")
ax.errorbar(x=df_glass['peak_pos'],
             y=df_glass['raw_cps'],
             yerr=df_glass['raw_cps_stdev'],
             fmt='x', color=[0.3]*3,
             lw=1)

ax.errorbar(x=df_glass_means.index,
            y=df_glass_means['raw_cps_mean'],
            # yerr=df_means['kraw_pcnt_stdev'],
            fmt='ok',
            markersize=8)

# Panel B: kraw for glass
ax = axs[1]
ax.set_ylabel("k-ratio %")
ax.set_title("Glass D2872 - background-corrected", fontsize=titlefontsize, loc="left")
ax.errorbar(x=df_glass['peak_pos'],
             y=df_glass['kraw_pcnt'],
             yerr=df_glass['kraw_pcnt'] * (df_glass['kraw_stdev_pcnt']/100),
             fmt='x', color=[0.3]*3,
             lw=1)

ax.errorbar(x=df_glass_means.index,
            y=df_glass_means['kraw_pcnt_mean'],
            # yerr=df_means['kraw_pcnt_stdev'],
            fmt='ok',
            markersize=8)

# Panel C: kraw for glass
ax = axs[2]
ax.set_ylabel("k-ratio %")
ax.set_title("Buddingtonite - background-corrected", fontsize=titlefontsize, loc="left")
ax.errorbar(x=df_budd['peak_pos'],
             y=df_budd['kraw_pcnt'],
             yerr=df_budd['kraw_pcnt'] * (df_budd['kraw_stdev_pcnt']/100),
             fmt='x', color=[0.3]*3,
             lw=1, label="Spot analysis")

ax.errorbar(x=df_budd_means.index,
            y=df_budd_means['kraw_pcnt_mean'],
            # yerr=df_means['kraw_pcnt_stdev'],
            fmt='ok',
            markersize=8, label="Mean")

ax.legend(fontsize=8, frameon=False, borderpad=0.2, edgecolor="w", handletextpad=0.2, markerscale=0.8)

# Annotations

plt.xlabel('Spectrometer position, L (mm)')
axs[2].set_xticks(np.arange(145, 149, 1))
# set minor ticks
axs[2].set_xticks(np.arange(145, 149, 0.2), minor=True)
axs[2].set_xlim(144.8, 148.2)

for ax in axs:
    ax.axvline(146.6, linestyle='--', zorder=0, lw=3, color='tab:blue', alpha=0.2)
    ax.axvline(147.0, linestyle='--', zorder=0, lw=3, color='tab:orange', alpha=0.2)

plt.sca(axs[0])
plt.annotate('GaN', 
             xy=(146.6, 36.7), xycoords='data', 
             xytext=(146.6, 37.3), textcoords='data', fontsize=10, color="tab:blue",
             arrowprops=dict(width=1, headwidth=7, headlength=5, facecolor='tab:blue', lw=1, alpha=0.5), 
             annotation_clip=False,
             ha='center')

plt.annotate('BN', 
             xy=(147.0, 36.7), xycoords='data', 
             xytext=(147.0, 37.3), textcoords='data', fontsize=10, color="tab:orange",
             arrowprops=dict(width=1, headwidth=7, headlength=5, facecolor='tab:orange', lw=1, alpha=0.5),  
             annotation_clip=False,
             ha='center')

letterlabel = "ABC"
trans = mtransforms.ScaledTranslation(-5/72, -5/72, fig.dpi_scale_trans)

for i, letter in enumerate(letterlabel):
    axs[i].text(1.0, 1.0, letter, transform=axs[i].transAxes + trans,
            fontsize=12, ha='right', va='top', fontfamily='Arial',
            bbox=dict(facecolor='w', edgecolor='grey', pad=3.0))

plt.tight_layout()
plt.savefig("../figures/peak_positions.png")
plt.savefig("../figures/for_submission/Figure_9.pdf", dpi=600)

plt.show()
```
